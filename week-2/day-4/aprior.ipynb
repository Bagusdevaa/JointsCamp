```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_excel('http://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx')
df.head()
```

```python
df.shape
```

```python
# Menghapus white space berlebih
df['Description'] = df['Description'].str.strip()
```

```python
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
df['InvoiceNo'] = df["InvoiceNo"].astype('str')
```

```python
df = df[~df['InvoiceNo'].str.contains("C")] 
```

```python
# Grouping
new_df = df.groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index("InvoiceNo")
```

```python
def one_hot(x):
    if x <=0:
        return 0
    if x>1:
        return 1
```

```python
basket = new_df.applymap(one_hot)
```

```python
basket
```

```python
frq_items = apriori(basket, min_support= 0.5, use_colnames= True)
```

```python
rules = association_rules(frq_items, metric='lift', min_threshold=1)
rules = rules.sort_values(['confidence','lift'], ascending=[False, False])
```
